```python
# linear regression dataset
import numpy as np
from numpy import matrix
from numpy import array
from matplotlib import pyplot
# define dataset
data = array([
	[0.05, 0.12],
	[0.18, 0.22],
	[0.31, 0.35],
	[0.42, 0.38],
	[0.5, 0.49]])
print(data)
# split into inputs and outputs
X, y = data[:,0], data[:,1]
print(X)
print(y)
X = X.reshape((len(X), 1))
print(X)
# scatter plot
pyplot.scatter(X, y)
pyplot.show()
```
